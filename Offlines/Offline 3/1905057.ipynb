```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split
from torchvision import datasets, transforms
from tqdm import tqdm

# Set random seed for reproducibility
np.random.seed(42)

# 1. Data Loading and Preprocessing using torchvision
def load_fashion_mnist():
    transform = transforms.Compose([transforms.ToTensor()])
    train_dataset = datasets.FashionMNIST(root='./data', train=True, download=True, transform=transform)
    test_dataset = datasets.FashionMNIST(root='./data', train=False, download=True, transform=transform)

    X_train = train_dataset.data.numpy().reshape(-1, 28*28) / 255.0  # normalize
    y_train = np.eye(10)[train_dataset.targets.numpy()]  # One-hot encode
    X_test = test_dataset.data.numpy().reshape(-1, 28*28) / 255.0
    y_test = np.eye(10)[test_dataset.targets.numpy()]

    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
    return X_train, y_train, X_val, y_val, X_test, y_test

X_train, y_train, X_val, y_val, X_test, y_test = load_fashion_mnist()

# 2. Improved Neural Network Components
class DenseLayer:
    def __init__(self, input_dim, output_dim):
        self.weights = np.random.randn(input_dim, output_dim) * np.sqrt(2.0 / input_dim)
        self.bias = np.zeros((1, output_dim))

    def forward(self, X):
        self.input = X
        return np.dot(X, self.weights) + self.bias

    def backward(self, d_out, learning_rate):
        d_weights = np.dot(self.input.T, d_out)
        d_bias = np.sum(d_out, axis=0, keepdims=True)
        d_input = np.dot(d_out, self.weights.T)

        self.weights -= learning_rate * d_weights
        self.bias -= learning_rate * d_bias
        return d_input

class ReLU:
    def forward(self, X):
        self.input = X
        return np.maximum(0, X)

    def backward(self, d_out):
        d_input = d_out.copy()
        d_input[self.input <= 0] = 0
        return d_input

class Softmax:
    def forward(self, X):
        exp_X = np.exp(X - np.max(X, axis=1, keepdims=True))
        return exp_X / np.sum(exp_X, axis=1, keepdims=True)

    def backward(self, predicted, target):
        return predicted - target

# Cross-entropy loss function
def cross_entropy_loss(predictions, targets):
    return -np.mean(np.sum(targets * np.log(predictions + 1e-8), axis=1))

# Define the neural network with selective backward call for each layer
class NeuralNetwork:
    def __init__(self, architecture):
        self.layers = architecture

    def forward(self, X):
        for layer in self.layers:
            X = layer.forward(X)
        return X

    def backward(self, predictions, targets, learning_rate):
        loss_grad = self.layers[-1].backward(predictions, targets)
        for layer in reversed(self.layers[:-1]):
            if 'learning_rate' in layer.backward.__code__.co_varnames:
                loss_grad = layer.backward(loss_grad, learning_rate)
            else:
                loss_grad = layer.backward(loss_grad)

# 3. Training Function with Mini-batch Gradient Descent
def train_model(model, X_train, y_train, X_val, y_val, epochs, learning_rate, batch_size=64):
    loss_history = []
    accuracy_history = []
    f1_history = []
    val_loss_history = []

    for epoch in range(epochs):
        # Training loop
        for i in range(0, X_train.shape[0], batch_size):
            X_batch = X_train[i:i + batch_size]
            y_batch = y_train[i:i + batch_size]

            predictions = model.forward(X_batch)
            loss = cross_entropy_loss(predictions, y_batch)
            model.backward(predictions, y_batch, learning_rate)

        # Validation metrics
        val_predictions = model.forward(X_val)
        val_loss = cross_entropy_loss(val_predictions, y_val)
        val_accuracy = accuracy_score(np.argmax(y_val, axis=1), np.argmax(val_predictions, axis=1))
        val_f1 = f1_score(np.argmax(y_val, axis=1), np.argmax(val_predictions, axis=1), average='macro')

        loss_history.append(loss)
        accuracy_history.append(val_accuracy)
        f1_history.append(val_f1)
        val_loss_history.append(val_loss)

        print(f"Epoch {epoch+1}, Loss: {loss:.4f}, Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}, F1 Score: {val_f1:.4f}")

    return loss_history, accuracy_history, f1_history, val_loss_history

# Define models with different architectures
models = [
    ("Model 1", [
        DenseLayer(28*28, 512), ReLU(), DenseLayer(512, 256), ReLU(),
        DenseLayer(256, 128), ReLU(), DenseLayer(128, 10), Softmax()
    ]),
    ("Model 2", [
        DenseLayer(28*28, 256), ReLU(), DenseLayer(256, 128), ReLU(),
        DenseLayer(128, 64), ReLU(), DenseLayer(64, 10), Softmax()
    ]),
    ("Model 3", [
        DenseLayer(28*28, 1024), ReLU(), DenseLayer(1024, 512), ReLU(),
        DenseLayer(512, 256), ReLU(), DenseLayer(256, 10), Softmax()
    ])
]

# Define learning rates
learning_rates = [0.005, 0.001, 0.0005, 0.0001]

# Initialize variables to track the best model
best_model_info = {"name": None, "learning_rate": None, "val_accuracy": 0}

# Training loop for all models and learning rates
epochs = 10
for model_name, architecture in models:
    for lr in learning_rates:
        print(f"Training {model_name} with learning rate: {lr}")
        model = NeuralNetwork(architecture)
        loss_history, accuracy_history, f1_history, val_loss_history = \
            train_model(model, X_train, y_train, X_val, y_val, epochs=epochs, learning_rate=lr)

        # Check if this model is the best based on validation accuracy
        if max(accuracy_history) > best_model_info["val_accuracy"]:
            best_model_info["name"] = model_name
            best_model_info["learning_rate"] = lr
            best_model_info["val_accuracy"] = max(accuracy_history)

        # Plot metrics
        epochs_range = range(1, epochs + 1)
        plt.figure(figsize=(12, 10))
        for i, (metric, history, label) in enumerate(
                zip(
                    ["Loss", "Validation Accuracy", "F1 Score"],
                    [loss_history, accuracy_history, f1_history],
                    ['Loss', 'Val Accuracy', 'F1 Score'])):
            plt.subplot(2, 2, i+1)
            plt.plot(epochs_range, history, label=f'{model_name} - LR: {lr}')
            plt.xlabel('Epoch')
            plt.ylabel(label)
            plt.legend()

        # Confusion Matrix Plot
        val_predictions = model.forward(X_val)
        val_predictions_classes = np.argmax(val_predictions, axis=1)
        y_val_classes = np.argmax(y_val, axis=1)
        conf_matrix = confusion_matrix(y_val_classes, val_predictions_classes)

        plt.subplot(2, 2, 4)
        plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
        plt.colorbar()
        plt.title("Confusion Matrix")
        plt.xlabel("Predicted Label")
        plt.ylabel("True Label")
        plt.tight_layout()

        plt.show()

# Print the best model's information
print(f"Best Model: {best_model_info['name']} with Learning Rate: {best_model_info['learning_rate']}")
print(f"Best Validation Accuracy: {best_model_info['val_accuracy']:.4f}")
```
